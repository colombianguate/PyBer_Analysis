```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
names = ['Tony', 'Joe', 'Sally', 'Andrea']
num_marbles = [100, 500, 150, 20]
```

```python
fig, ax = plt.subplots(2,2, figsize=(20,16))
ax[0,0].plot(names, num_marbles)
ax[0, 0].set_title("Axis [0,0]")
ax[0,0].set_xlabel("names")
ax[0,1].bar(names, num_marbles)
ax[0, 1].set_title("Axis [0,1]")
ax[1,0].scatter(names, num_marbles, s=num_marbles)
ax[1, 0].set_title("Axis [1,0]")
ax[1,1].pie(num_marbles, labels = names)
ax[1, 1].set_title("Axis [1,1]")

#show
plt.show()
plt.subplots_adjust(hspace=.2)
```
